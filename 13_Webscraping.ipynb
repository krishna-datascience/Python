## Scrapping Samsung Phones

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
my_url = 'https://www.flipkart.com/search?p%5B%5D=facets.brand%255B%255D%3DSamsung&sid=tyy%2F4io&sort=recency_desc&wid=1.productCard.PMU_V2_1'
```

```python
ureq = urlopen(my_url)
page_html = ureq.read()
ureq.close()
```

```python
page_soup = BeautifulSoup(page_html,'html.parser')
```

```python
Mobiles = page_soup.findAll('div',{'class':'_3O0U0u'})
```

```python
print(len(Mobiles))
```

```python
for Mobile in Mobiles:
    name = Mobile.findAll('div',{'class':'_3wU53n'})
    name = name[0].text
    name = name.split(')')
    final_name = name[0] + ')'
    final_name = final_name.replace(',',' |')

    rating = Mobile.findAll('div',{'class':'hGSR34'})
    if rating:
        final_rating = rating[0].text
    else:
        final_rating = '-'
    
    price = Mobile.findAll('div',{'class':'_1vC4OE _2rQ-NK'})
    if price:
        price = price[0].text
        price = price.replace(',','')
        final_price = price.replace('₹','')
    else:
        final_price ='-'

    print(final_name,final_rating,final_price)
```

```python
mobile_list = [('Name','Rating','Price')]
for Mobile in Mobiles:
    name = Mobile.findAll('div',{'class':'_3wU53n'})
    name = name[0].text
    name = name.split(')')
    final_name = name[0] + ')'
    final_name = final_name.replace(',',' |')

    rating = Mobile.findAll('div',{'class':'hGSR34'})
    if rating:
        final_rating = rating[0].text
    else:
        final_rating = '-'
    
    price = Mobile.findAll('div',{'class':'_1vC4OE _2rQ-NK'})
    if price:
        price = price[0].text
        price = price.replace(',','')
        final_price = price.replace('₹','')
    else:
        final_price ='-'

    mobile_list.append((final_name,final_rating,final_price))
    
from tabulate import tabulate
print(tabulate(mobile_list,tablefmt="pipe"))
```

## Another Example of Scraping -> Codechef Contest page scrapping

```python
import requests
from bs4 import BeautifulSoup
from pprint import pprint
```

```python
url = "https://www.codechef.com/contests"
r = requests.get(url)
soup = BeautifulSoup(r.content)
eles = soup.findAll('table', attrs={'class': 'dataTable'})
```

```python
data = []
for i in eles:
    v = i.findAll('tr')
    d = []
    for j in range(len(v)):
        if j == 0:
            x = v[j].findAll('th')
        else:
            x = v[j].findAll('td')
        d.append([i.text.strip() for i in x])
    data.append(d)
pprint(data)
```

```python
for i in data[:2]:
    print(tabulate(i,tablefmt='fancy_grid'))
    print()
```

## Another Example - > Cricinfo Scrapping

```python
import requests
from bs4 import BeautifulSoup
from pprint import pprint
from tabulate import tabulate
import csv
```

```python
url = "http://www.espncricinfo.com/rankings/content/page/211271.html"
r = requests.get(url)
soup = BeautifulSoup(r.content)
headings = soup.findAll('h3')
headings = [i.text.strip() for i in headings]
pprint(headings)
```

```python
tables = soup.findAll('table', attrs={'class': 'StoryengineTable'})
print(len(tables))
print(tables)
```

```python
data = []
for i in tables:
    v = i.findAll('tr')
    d = []
    for j in range(len(v)):
        if j == 0:
            x = v[j].findAll('th')
        else:
            x = v[j].findAll('td')
        d.append([i.text.strip() for i in x])
    data.append(d)
pprint(data)
```

```python
for i in range(len(data)):
    print(headings[i])
    print(tabulate(data[i],tablefmt='fancy_grid'))
    print()
```

```python
for i in range(len(data)):
    filename = "ICC_Rankings"+str(i+1)+".csv"
    with open('./data/'+filename, "w") as f:
        f.write(headings[i]+"\n")
        writer = csv.writer(f)
        writer.writerows(data[i])
```

### using Robobrowser

```python
import re
from robobrowser import RoboBrowser

from bs4 import BeautifulSoup as bs

def getData(from_station, to_station):
    br=RoboBrowser(history=True)
    br.open("http://www.delhimetrorail.com/metro-fares.aspx")
    form=br.get_form()
    form['ctl00$MainContent$ddlFrom'].value=from_station
    form['ctl00$MainContent$ddlTo'].value=to_station
    br.submit_form(form,submit=form['ctl00$MainContent$btnShowFare'])
    src=str(br.parsed())
    return src

if __name__ == '__main__':

    st=getData("HUDA CITY CENTRE","ANAND VIHAR")
    nf= st.index('class="fare_new_nor_right"')
    print("Normal Fare: ",st[nf+27:nf+29])

    cf=st.index('class="fare_new_right_right"')
    print("Concessional Fare: ", st[cf + 29:cf + 31])
    tm = st.index('Timing -')
    tm1=st[tm + 9:tm + 13]
    for s in tm1.split():
        if s.isdigit():
            print("Timing: ",s)


    stn = st.index('Stations -')
    print("Stations: ", st[stn + 11:stn + 13])

    inter = st.index('Interchange -')
    print("Interchange: ", st[inter + 14:inter + 15])
```
