```python
a = get_close_matches('rainn',['train','rain','again','rainy','python'],n=4)
print(a)
b = get_close_matches('rainn',list(data.keys()),n=15,cutoff=0.86)
print(b)

b = get_close_matches('frnd',list(data.keys()),cutoff=0.6)
print(b)
```

```python
def meaning(word):
    
    if word in data:
        print(data[word])
        
    elif len(get_close_matches(word,list(data.keys())))>0:
        a = get_close_matches(word,list(data.keys()))[0]
        print(data[a])
    else:
        print("Word doesn't Exist")
        
usr = input("Enter a word : ").lower()
meaning(usr)
```

```python
def meaning(word):
    
    if word in data:
        print(data[word])
        
    elif len(get_close_matches(word,list(data.keys())))>0:
        a = get_close_matches(word,list(data.keys()))
        for i,v in enumerate(a,1):
            print(i,'->',v)
        n = int(input("Enter your Choice (1-3) : "))
        
        print(data[a[n-1]])
    else:
        print("Word doesn't Exist")
        
usr = input("Enter a word : ").lower()
meaning(usr)
```

```python
import json
from difflib import get_close_matches
data = json.load(open("data.json"))

def meaning(word):
    word = word.lower()
    if word in data:
        return ("THe meaning of word {} is : \n{}".format(word,data[word]))
    
    elif word.upper() in data:
        word = word.upper()
        return ("THe meaning of word {} is : \n{}".format(word,data[word]))
    elif word.title() in data:
        word = word.title()
        return("THe meaning of word {} is : \n{}".format(word,data[word]))
    elif len(get_close_matches(word,list(data.keys()))) > 0:
        word = get_close_matches(word,list(data.keys()),5)
        
        for n,w in enumerate(word,1):
            print(n,":->",w)
        print("Please enter the number for the word you want")
        while True:
            try:
                a = int(input())
                if a>5 or a<= 0:
                    print("Please enter a number in the range")
                    continue
                break
            except:
                print("Please Enter a number")
        word = word[a-1]
        return("THe meaning of word {} is : \n{}".format(word,data[word]))
        
    else:
        return("WORD NOT FOUND")
        
while True:
    print("Press 1 for closing this application")
    print("Press 2 for checking the meaning of a word")
    n = int(input())
    if n == 1:
        break
    else:
        word = input("Enter the Word : ")
        m = meaning(word)
        print(m)
```
